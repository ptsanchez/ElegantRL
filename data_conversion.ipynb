```python
import numpy as np
import pandas as pd
import os
import time
import psutil

# --- Configuration ---
ticker = 'AAPL'
data_dir = './data'
feature_path = os.path.join(data_dir, f"{ticker}.npy")
time_path = os.path.join(data_dir, f"{ticker}_time.npy")

# Parameters for the test
N_TRIALS = 10
EPISODE_WINDOW = 256  # Simulating an agent looking at last 64 bars
CLEAR_CACHE_WARNING = True

def get_process_memory():
    process = psutil.Process(os.getpid())
    return process.memory_info().rss / 1024 / 1024  # MB

def benchmark_sampling(n_trials=10, window_size=64):
    print(f"--- Benchmarking: {ticker} ---")
    print(f"Window Size: {window_size} rows")
    print(f"Initial RAM Usage: {get_process_memory():.2f} MB")
    
    # 1. Open File in MMAP mode (Zero copy, nearly instant)
    # We do this OUTSIDE the loop because in your RL env, you open it once in __init__
    start_load = time.time()
    features_mmap = np.load(feature_path, mmap_mode='r')
    times_mmap = np.load(time_path, mmap_mode='r')
    print(f"File Open Time: {(time.time() - start_load)*1000:.4f} ms")
    
    max_idx = len(features_mmap) - window_size
    timings = []
    
    print(f"\nRunning {n_trials} random lookups...")
    print("-" * 60)
    print(f"{'Trial':<6} | {'Index':<10} | {'Read Time (ms)':<15} | {'Date Found'}")
    print("-" * 60)

    for i in range(n_trials):
        # A. Pick Random Start Index
        rand_idx = np.random.randint(0, max_idx)
        
        # B. Start Timer
        t0 = time.perf_counter()
        
        # C. The Critical Operation: Slicing + Copying
        # mmap returns a view. .copy() forces the OS to actually read the disk/cache
        obs_slice = features_mmap[rand_idx : rand_idx + window_size].copy()
        time_val = times_mmap[rand_idx] # Just reading the int64 timestamp
        
        # D. Stop Timer
        t1 = time.perf_counter()
        dt_ms = (t1 - t0) * 1000
        timings.append(dt_ms)
        
        # For display, convert timestamp
        time_str = str(pd.to_datetime(time_val))
        
        print(f"{i+1:<6} | {rand_idx:<10} | {dt_ms:<15.4f} | {time_str}")

    # --- Statistics ---
    avg_time = np.mean(timings)
    std_time = np.std(timings)
    min_time = np.min(timings)
    max_time = np.max(timings)
    
    print("-" * 60)
    print(f"\nFinal Statistics ({n_trials} trials):")
    print(f"Average Read Time: {avg_time:.4f} ms")
    print(f"Std Dev:           {std_time:.4f} ms")
    print(f"Fastest Read:      {min_time:.4f} ms")
    print(f"Slowest Read:      {max_time:.4f} ms")
    print(f"Final RAM Usage:   {get_process_memory():.2f} MB (Should be low!)")

    # Throughput Estimation
    # How many steps per second could the data loader theoretically support?
    steps_per_sec = 1000 / avg_time

    print(f"\nTheoretical Throughput: ~{int(steps_per_sec)} steps/second")

if __name__ == "__main__":
    if os.path.exists(feature_path):
        benchmark_sampling(N_TRIALS, EPISODE_WINDOW)
    else:
        print(f"File not found: {feature_path}. Run the conversion script first.")
```

```python
import pandas as pd
import numpy as np
import os
import json
from tqdm import tqdm

# --- Configuration ---
source_dir = '/opt/rws/repos/RWS_LightGBM/data/reprocessed/final_merged_with_spy/ml_ready_with_targets'
dest_dir = './data'
REFERENCE_TICKER = "AAPL" 
cols_to_drop = ['ticker', 'ticker_d1'] 

os.makedirs(dest_dir, exist_ok=True)
files = [f for f in os.listdir(source_dir) if f.endswith('.parquet')]

# ---------------------------------------------------------
# Step A: Establish the "Gold Standard" Schema
# ---------------------------------------------------------
print(f"Loading reference ticker ({REFERENCE_TICKER}) to establish schema...")
ref_path = os.path.join(source_dir, f"{REFERENCE_TICKER}.parquet")

if not os.path.exists(ref_path):
    raise FileNotFoundError(f"Reference ticker {REFERENCE_TICKER} not found in source!")

ref_df = pd.read_parquet(ref_path)

if 'timestamp' in ref_df.columns:
    ref_df = ref_df.drop(columns=['timestamp'])
ref_df = ref_df.drop(columns=[c for c in cols_to_drop if c in ref_df.columns], errors='ignore')

MASTER_FEATURES = ref_df.columns.tolist()
EXPECTED_COUNT = len(MASTER_FEATURES)

print(f"Standard Schema Established: {EXPECTED_COUNT} features.")

# Save Metadata
metadata = {
    "feature_names": MASTER_FEATURES,
    "num_features": EXPECTED_COUNT,
    "reference_ticker": REFERENCE_TICKER
}
with open(os.path.join(dest_dir, 'master_metadata.json'), 'w') as f:
    json.dump(metadata, f, indent=4)

# ---------------------------------------------------------
# Step B: Process All Files
# ---------------------------------------------------------
skipped_tickers = []
processed_count = 0

for file in tqdm(files, desc="Standardizing Data"):
    try:
        ticker_name = file.replace('.parquet', '')
        file_path = os.path.join(source_dir, file)
        
        # 1. Read
        df = pd.read_parquet(file_path)
        
        # 2. Extract Timestamp (BUT DO NOT SAVE YET)
        timestamps_array = None
        if 'timestamp' in df.columns:
            # Convert to correct units and store in memory variable
            timestamps_array = df['timestamp'].astype('datetime64[ns]').astype(np.int64).values
            # Drop from dataframe so it doesn't interfere with feature processing
            df = df.drop(columns=['timestamp'])
        
        # 3. Enforce Feature Consistency
        try:
            # Attempt to select exactly the columns we need
            df_clean = df[MASTER_FEATURES]
        except KeyError as e:
            # If columns are missing, we skip.
            # CRITICAL CHANGE: We haven't saved the time file yet, so nothing to clean up.
            missing_cols = list(set(MASTER_FEATURES) - set(df.columns))
            skipped_tickers.append({
                "ticker": ticker_name, 
                "reason": "Missing Columns",
                "missing_count": len(missing_cols),
                "example_missing": missing_cols[:3]
            })
            continue

        # 4. Convert Features
        features_array = df_clean.apply(pd.to_numeric, errors='coerce').astype(np.float32).values
        
        # Double check shape
        if features_array.shape[1] != EXPECTED_COUNT:
            skipped_tickers.append({"ticker": ticker_name, "reason": "Shape Mismatch"})
            continue
            
        # 5. FINAL SAVE (Atomic-ish)
        # Now that we know features are valid, we save both files.
        np.save(os.path.join(dest_dir, f"{ticker_name}.npy"), features_array)
        
        if timestamps_array is not None:
            np.save(os.path.join(dest_dir, f"{ticker_name}_time.npy"), timestamps_array)
            
        processed_count += 1
        
    except Exception as e:
        print(f"Error on {file}: {e}")

# ---------------------------------------------------------
# Step C: Report
# ---------------------------------------------------------
print("\n" + "="*40)
print(f"Processing Complete.")
print(f"Successfully Converted: {processed_count}")
print(f"Skipped Tickers:      {len(skipped_tickers)}")
print("="*40)

if skipped_tickers:
    print("\nSkipped Ticker Details (Outliers):")
    for item in skipped_tickers[:10]:
        print(f"{item['ticker']:<8} | {item['reason']}")

with open(os.path.join(dest_dir, 'skipped_report.json'), 'w') as f:
    json.dump(skipped_tickers, f, indent=4)
```

```python
import random

# --- Configuration ---
data_dir = './data'
NUM_TICKERS_TO_CHECK = 100   # How many different tickers to test
SAMPLES_PER_TICKER = 1     # How many random rows to read per ticker

def verify_dataset_integrity():
    print(f"--- Starting Global Integrity Check ---")
    print(f"Directory: {data_dir}")
    
    # 1. Discover all tickers based on the time files
    # We look for '_time.npy' to ensure we only check completed pairs
    all_files = os.listdir(data_dir)
    time_files = [f for f in all_files if f.endswith('_time.npy')]
    
    if not time_files:
        print("ERROR: No data files found!")
        return

    # Extract ticker names (remove '_time.npy')
    all_tickers = [f.replace('_time.npy', '') for f in time_files]
    print(f"Found {len(all_tickers)} total tickers.")
    
    # 2. Randomly select tickers to test
    selected_tickers = random.sample(all_tickers, min(NUM_TICKERS_TO_CHECK, len(all_tickers)))
    
    print(f"\nChecking {len(selected_tickers)} random tickers...")
    print("=" * 80)
    print(f"{'Ticker':<8} | {'Shape (Rows, Cols)':<20} | {'Sample Time':<22} | {'Feat[0]':<8} | {'Status'}")
    print("=" * 80)

    for ticker in selected_tickers:
        try:
            # Paths
            feature_path = os.path.join(data_dir, f"{ticker}.npy")
            time_path = os.path.join(data_dir, f"{ticker}_time.npy")
            
            # Check if pair exists
            if not os.path.exists(feature_path):
                print(f"{ticker:<8} | {'MISSING FEATURE FILE':<20} | {'---':<22} | {'---':<8} | FAIL")
                continue

            # 3. Load via Memory Map (Instant)
            features = np.load(feature_path, mmap_mode='r')
            times = np.load(time_path, mmap_mode='r')
            
            # 4. Consistency Checks
            if features.shape[0] != times.shape[0]:
                print(f"{ticker:<8} | {str(features.shape):<20} | {'MISMATCH':<22} | {'---':<8} | FAIL (Shape)")
                continue
                
            # 5. Random Sampling
            for _ in range(SAMPLES_PER_TICKER):
                # Pick random index
                idx = np.random.randint(0, len(features))
                
                # Read Data
                row_feats = features[idx]
                row_time_int = times[idx]
                
                # Decode Timestamp
                # Force unit='ns' to match our save format
                ts = pd.to_datetime(row_time_int, unit='ns')
                
                # Logic Check: Is the year valid? (Should be > 1980)
                if ts.year < 1980:
                    status = "FAIL (1970)"
                else:
                    status = "OK"

                # Print Row
                print(f"{ticker:<8} | {str(features.shape):<20} | {str(ts):<22} | {row_feats[0]:<8.3f} | {status}")
                
        except Exception as e:
            print(f"{ticker:<8} | {'ERROR':<20} | {str(e):<22} | {'---':<8} | CRASH")

    print("=" * 80)
    print("Check complete.")

if __name__ == "__main__":
    verify_dataset_integrity()
```

```python

# --- Configuration ---
source_dir = '/opt/rws/repos/RWS_LightGBM/data/reprocessed/final_merged_with_spy/ml_ready_with_targets'
REFERENCE_TICKER = "EOG" 

# Metadata columns to ignore during comparison
cols_to_ignore = ['ticker', 'ticker_d1', 'timestamp'] 

# ---------------------------------------------------------
# Step 1: Establish the "Gold Standard" from Reference
# ---------------------------------------------------------
ref_path = os.path.join(source_dir, f"{REFERENCE_TICKER}.parquet")

if not os.path.exists(ref_path):
    raise FileNotFoundError(f"Reference ticker {REFERENCE_TICKER} not found!")

# Read just the columns (efficient)
ref_df = pd.read_parquet(ref_path)
ref_cols = [c for c in ref_df.columns if c not in cols_to_ignore]
ref_set = set(ref_cols)

print(f"--- Schema Standard ({REFERENCE_TICKER}) ---")
print(f"Feature Count: {len(ref_cols)}")
print(f"Columns (First 5): {ref_cols[:5]}\n")

# ---------------------------------------------------------
# Step 2: Audit All Tickers
# ---------------------------------------------------------
files = [f for f in os.listdir(source_dir) if f.endswith('.parquet')]
skipped_log = []
valid_count = 0

print(f"Auditing {len(files)} files...")

for file in tqdm(files, desc="Checking Schemas"):
    try:
        # Read columns only
        # Note: pandas usually reads the whole file, but for just checking cols it's fast enough here.
        # If very slow, we could use pyarrow.parquet.read_schema()
        df = pd.read_parquet(os.path.join(source_dir, file))
        
        # Get feature columns for this ticker
        current_cols = [c for c in df.columns if c not in cols_to_ignore]
        current_set = set(current_cols)
        
        # CHECK 1: Feature Count
        if len(current_cols) != len(ref_cols):
            skipped_log.append({
                "ticker": file,
                "reason": "Count Mismatch",
                "details": f"Has {len(current_cols)} features (Expected {len(ref_cols)})"
            })
            continue

        # CHECK 2: Missing Specific Columns
        # We check if the Reference Set is a subset of Current Set
        # (This ensures all required EOG columns exist in the file)
        missing_cols = ref_set - current_set
        
        if missing_cols:
            skipped_log.append({
                "ticker": file,
                "reason": "Missing Required Features",
                "details": f"Missing {len(missing_cols)} cols: {list(missing_cols)[:3]}..."
            })
            continue
            
        valid_count += 1

    except Exception as e:
        skipped_log.append({"ticker": file, "reason": "Read Error", "details": str(e)})

# ---------------------------------------------------------
# Step 3: Final Report
# ---------------------------------------------------------
print("\n" + "="*50)
print(f"AUDIT COMPLETE")
print("="*50)
print(f"Valid Tickers:  {valid_count}")
print(f"Skipped Tickers: {len(skipped_log)}")
print("-" * 50)

if skipped_log:
    print(f"{'Ticker':<25} | {'Reason':<20} | {'Details'}")
    print("-" * 80)
    for item in skipped_log:
        print(f"{item['ticker']:<25} | {item['reason']:<20} | {item['details']}")
else:
    print("Perfect! All tickers match the schema.")
```
