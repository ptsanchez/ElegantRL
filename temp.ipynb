```python
import os
import pandas as pd
import random
from pathlib import Path
import datetime

# 1. Setup paths
# Replace with the actual path to your parquet files

# 2. Identify all ticker names in the folder
# This scans for *.parquet and removes the extension
tickers = [f.replace('.parquet', '') for f in os.listdir(folder_path) if f.endswith('.parquet')]

# 3. Your sampling logic
# Ensure we don't try to sample more tickers than we actually have
sample_size = min(len(tickers), 5)
sample_tickers = random.sample(tickers, sample_size)
random_samples = []


for ticker in sample_tickers:
    ticker_path = os.path.join(folder_path, f"{ticker}.parquet")
    df = pd.read_parquet(ticker_path)
    
    if not df.empty:
        # 1. Ensure the index or timestamp column is in datetime format
        if not isinstance(df.index, pd.DatetimeIndex):
            # If the timestamp is a column, set it as index first
            if 'timestamp' in df.columns:
                df.index = pd.to_datetime(df['timestamp'])
            else:
                df.index = pd.to_datetime(df.index)

        # 2. Filter for exactly 9:30 AM
        # This assumes the data is already in New York time
        df_at_open = df[df.index.time == datetime.time(9, 30)]
        
        # 3. Sample from the filtered subset
        if not df_at_open.empty:
            random_samples.append(df_at_open.sample(n=1))

# 4. Combine and display results
if random_samples:
    df_samples = pd.concat(random_samples)
    display(df_samples)
else:
    print("No market open data found to sample.")
```

```python
import os
import pandas as pd

data_path = '/opt/rws/repos/RWS_LightGBM/data/reprocessed/final_merged_with_spy/ml_ready_with_targets'
tickers = [f for f in os.listdir(data_path) if f.endswith('.parquet')]

print(f"Scanning {len(tickers)} tickers...")

# 1. Get columns from the first ticker as a baseline
first_ticker = 'AAPL.parquet'
first_cols = set(pd.read_parquet(os.path.join(data_path, first_ticker)).columns)
common_columns = first_cols.copy()

# 2. Find the intersection (columns that exist in EVERY file)
diff_found = False
for ticker in tickers:
    # Use columns=[] to read metadata only (much faster)
    current_cols = set(pd.read_parquet(os.path.join(data_path, ticker)).columns)
    
    # Check if this ticker is missing anything from our baseline
    missing = common_columns - current_cols
    if missing:
        print(f"⚠️ {ticker} is missing {len(missing)} columns: {list(missing)}...")
        common_columns &= current_cols
        diff_found = True

if not diff_found:
    print("✅ All tickers have identical columns.")
else:
    print(f"\nSummary:")
    print(f"- Total unique columns in first file: {len(first_cols)}")
    print(f"- Total common columns across all files: {len(common_columns)}")
    print(f"- Columns to be excluded: {len(first_cols - common_columns)}")
```
